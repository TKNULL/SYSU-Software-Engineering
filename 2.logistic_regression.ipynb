## 映射

找一个映射到(0, 1)区间的假设

$h_\theta(x) = g(\theta^Tx)$

$g(z) = \frac{1}{1 + e^{-z}}$

得到的是个概率：$P(y = 0 \space |\space x; \theta)$

```python
g = lambda x : 1 / (1 + (math.e ** (-x)))
x = np.linspace(-10, 10, 1000)
y = np.vectorize(g)(x)

plt.plot(x, y, color='b', label="g(z)")
plt.legend(loc="upper left")
```
